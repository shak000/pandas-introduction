# Pandas datastructure
    1. Series (1-Dimensional). 
    2. Dataframe (2-Dimensional).
    3. Panel (Multi-Dimensional).

```python
# First import panda
import pandas as pd
```

```python
# Creating an empty series.
pd.Series()
```

```python
# Creating series with list
my_list = ["Processor", "GPU", "RAM"]
pd.Series(my_list)
```

```python
# you can change index name
pd.Series(my_list, index = ["a", "b", "c"])
```

```python
# Creating series with dictionary
my_dict = {"name" : "Shakil", "Age" : 26}
pd.Series(my_dict)
```

```python
# creating series with numpy array
import numpy as np
pc_component = np.array(["PSU", "Motherboard", "Monitor"])
pd.Series(pc_component)
```

```python
# Creating Dataframe
pc_brand = np.array(["Corsair", "Gigabyte", "LG"])
pd.DataFrame({"PC Componenst" : pc_component, "Brand" : pc_brand}, index = [1,2,3])
```
